```python
from ms_pred.mabnet import mol_pretrain_data
import pandas as pd
from torch.utils.data import DataLoader

tree_processor = mol_pretrain_data.TreeProcessor(
    pe_embed_k=14,
    root_encode='egt2d',
    binned_targs=False,
    add_hs=True,
)

df = pd.read_pickle('/home/weiwentao/workspace/ms-pred/data/spec_datasets/canopus_train_public/labels_egt.pkl')

dataset = mol_pretrain_data.PretrainDataset(
    df=df[:20],
    tree_processor=tree_processor,
)

collate_fn = dataset.get_collate_fn()
train_loader = DataLoader(
        dataset,
        num_workers=min(12, 8),
        collate_fn=collate_fn,
        shuffle=True,
        batch_size=8,
        persistent_workers=12,
    )
```

```python
a = next(iter(train_loader))  # This will trigger the collate_fn and process the data
```
